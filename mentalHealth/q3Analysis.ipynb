<h1>How do gender and other demographic factors influence attitudes towards mental health in the workplace?</h1>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', 200)
pd.set_option('display.max_colwidth', 200)

df = pd.read_csv("./Data/cleanedData.csv")
#questions used
#IDs: [1, 2, 3, 4, 5, 10, 11, 14, 20, 27, 50, 51, 55]
```

```python
age_brackets = pd.cut(df[df['QuestionID'] == 1]['AnswerText'].astype(int), bins=range(0, 101, 5))
age_brackets.value_counts().sort_index().plot(kind='bar')

plt.xlabel('Age Bracket')
plt.ylabel('Count')
plt.title('What is your age?')
plt.show()
```

```python

df['AnswerText'] = df['AnswerText'].str.lower()
df['AnswerText'] = df['AnswerText'].replace('Male', 'male')
df['AnswerText'] = df['AnswerText'].replace('Female', 'female')
df['AnswerText'] = df['AnswerText'].replace('nonbinary', 'non-binary')

# Filter out -1 values
df_filtered = df[df['QuestionID'] == 2]
df_filtered = df_filtered[df_filtered['AnswerText'] != '-1']

# Get the counts of each answer
answer_counts = df_filtered['AnswerText'].value_counts()

# Get the top 5 answers with the highest count
top_answers = answer_counts.head(5)

# Get the count of all other answers
other_count = answer_counts.sum() - top_answers.sum()

# Create a new dataframe for the chart data
chart_data = pd.DataFrame({'Answer': top_answers.index.tolist() + ['Other'], 'Count': top_answers.tolist() + [other_count]})

# Plot the chart
chart_data.plot(x='Answer', y='Count', kind='bar')



plt.xlabel('Answer')
plt.ylabel('Count')
plt.title('Top 5 Answers with Most Count')

plt.show()

chart_data.plot(x='Answer', y='Count', kind='pie', autopct='%1.1f%%')
plt.show()
```
